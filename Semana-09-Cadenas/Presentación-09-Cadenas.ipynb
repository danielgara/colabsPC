**Obtener el tamaño de una cadena – len(mensaje)**

```python
mensaje = "Hola, Grupo"
saludo = "Hola"
tamaño1 = len(mensaje)
tamaño2 = len(saludo)
tamaño3 = len(mensaje+saludo)

print(tamaño1)
print(tamaño2)
print(tamaño3)
```

**Extraer un carácter de una cadena – mensaje[pos]**

```python
mensaje = "Hola, Grupo"
pos0 = mensaje[0]
pos4 = mensaje[4]
posFinal = mensaje[len(mensaje)-1]

print(pos0)
print(pos4)
print(posFinal)
```

**Buscar un carácter o subcadena en una cadena – mensaje.find(“textoABuscar”)**

```python
mensaje = "Hola, Grupo"
saludo = "Hola"
buscar1 = mensaje.find("G")
buscar2 = mensaje.find("Globo")
buscar3 = mensaje.find(saludo)

print(buscar1)
print(buscar2)
print(buscar3)
```

**Buscar cadena en una cadena basado en posición inicial – mensaje.find(“textoABuscar”, posInicial)**

```python
mensaje = "Hola, Hola"
saludo = "Hola"
buscar1 = mensaje.find("o")
buscar2 = mensaje.find("o", buscar1)
buscar3 = mensaje.find("o", buscar1+1)
buscar4 = mensaje.find(saludo, 4)

print(buscar1)
print(buscar2)
print(buscar3)
print(buscar4)
```

**Strings con ciclos**

```python
mensaje = input("Ingrese un texto: ")

i = 0
while(i < len(mensaje)):
  print(mensaje[i])
  i = i+1

print("Fin programa")
```

**Otros métodos – análisis**

```python
mensaje = "Hola, Grupo"

# devuelve la cadena en minúscula 
mensajeMinus = mensaje.lower()
print(mensajeMinus)

# devuelve la cadena en mayúscula
mensajeMayus = mensaje.upper()
print(mensajeMayus)

# cuenta cuantas veces aparece la "o" en mensaje
conteo = mensaje.count("o")
print(conteo)

# devuelve la cadena reemplazando las "o" por "u"
nuevoMensaje = mensaje.replace("o","u")
print(nuevoMensaje)
```
