**Condicional if-then-else**

```python
temperatura = float(input("Ingrese la temperatura: "))

if (temperatura > 27):
  print("Comprar helado")
else:
  print("Comprar jugo de naranja")

print("Fin programa")
```

**Condicional if-then-else – Ejercicio 1**

```python
ladoCuadrado = float(input("Ingrese el lado del cuadrado: "))

if(ladoCuadrado > 0):
  areaCuadrado = ladoCuadrado*ladoCuadrado
  print("El area es: "+str(areaCuadrado))
else:
  print("Lado no puede ser menor o igual a cero")

print("Fin programa")
```

**Ejercicio de clase 1**

```python
votosValidos = int(input("Ingrese votos válidos: "))
votosPartido = int(input("Ingrese votos por su partido: "))

umbral = votosValidos*0.03

if(votosPartido > umbral):
  print("Tu partido tendrá curules")
else:
  print("Se quemaron")
```

**Sección elif**

```python
temperatura = float(input("Ingrese la temperatura: "))

if (temperatura > 27):
  print("Comprar helado")
elif (temperatura < 15):
  print("Comprar chocolate")
else: 
  print("Comprar jugo de naranja")
  
print("Fin programa")
```

**Condicional con elif – Ejercicio 2**

```python
cantidadFlores = int(input("Ingrese la cantidad de flores: "))

if (cantidadFlores >= 10):
  precioTotal = cantidadFlores*8
  print("Total a pagar: " +str(precioTotal))
elif (cantidadFlores >= 5):
  precioTotal = cantidadFlores*10
  print("Total a pagar: " +str(precioTotal))
elif (cantidadFlores == 3):
  print("No debes pagar nada, las flores son gratis")
else:
  precioTotal = cantidadFlores*15
  print("Total a pagar: " +str(precioTotal))
  
print("Fin programa")
```

**Condicional con elif – Ejercicio 3**

```python
edad = int(input("Ingrese su edad: "))

if (edad >= 18):
  print("Mayor de edad")
elif (edad >= 60):
  print("Adulto mayor")
```

**Ejercicio de clase 2**

```python
año = int(input("Ingrese año de nacimiento: "))

if(año >= 1994 and año <= 2010):
  print("Eres Generación Z")
elif(año >= 1981 and año <= 1993):
  print("Eres Milennial")
else:
  print("Eres de otra generación")
```

**Condicionales anidados – Código**

```python
edad = int(input("Ingrese su edad: "))

if (edad >= 18):
  dinero = float(input("Ingrese su dinero: "))
  if (dinero >= 10):
    print("Puedes ingresar al teatro")
  else:
    print("No cumples con el dinero")
else:
  print("No cumples con la edad")

print("Fin programa")
```
