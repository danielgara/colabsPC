**Método plot**

```python
import matplotlib.pyplot as plt

#Imaginemos que x es el tiempo en años
x = [0, 1, 2, 3, 4, 5]

#Imaginemos que y es el dinero total ahorrado
y = [0, 50, 70, 70, 100, 246]

plt.plot(x, y)
plt.xlabel("Tiempo (años)")
plt.ylabel("Ahorros (dólares)")

plt.show()
```

**Método bar**

```python
import matplotlib.pyplot as plt

#Imaginemos que x son los años
x = [2000, 2001, 2002, 2003, 2004, 2005]

#Imaginemos que y es la cantidad de canciones
#Lanzadas ese año
y = [5, 2, 0, 8, 3, 5]

plt.bar(x, y)
plt.xlabel("Año")
plt.ylabel("Canciones lanzadas")

plt.show()
```
