**Ejercicio 3: Solución errónea helado-malteada**

```python
temperatura = float(input("Ingrese la temperatura: "))

if(temperatura > 27):
  print("Comprar helado")

print("Comprar malteada")
print("Fin programa")
```

**Ejercicio 3: Solución correcta helado-malteada**

```python
temperatura = float(input("Ingrese la temperatura: "))

if(temperatura > 27):
  print("Comprar helado")
  print("Comprar malteada")

print("Fin programa")
```
