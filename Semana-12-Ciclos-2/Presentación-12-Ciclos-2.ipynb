**Contadores – Listas simples**

```python
edades = [5,20,35,16,12]

contadorMenores = 0
i = 0
while(i < len(edades)):
  if(edades[i] <= 17):
    contadorMenores = contadorMenores+1
  i = i+1

print("Total menores: "+str(contadorMenores))
```

**Contadores - solución**

```python
edades = [5,20,35,16,12]

contadorMenores = 0
contadorMayores = 0
i = 0
while(i < len(edades)):
  if(edades[i] <= 17):
    contadorMenores = contadorMenores+1
  elif(edades[i] >= 18):
    contadorMayores = contadorMayores+1
  i = i+1

print("Total menores: "+str(contadorMenores))
print("Total mayores: "+str(contadorMayores))
```

**Contadores – Ejemplo 2 - Solución**

```python
maquina1 = {"nombre":"Tractor", "precio":500}
maquina2 = {"nombre":"Fumigadora", "precio":120}
maquina3 = {"nombre":"Cosechadora", "precio":1060}
maquina4 = {"nombre":"Aplanadora", "precio":730}
listaMaquinas = [maquina1, maquina2, maquina3, maquina4]

contadorBaratas = 0
contadorCaras = 0
i = 0
while(i < len(listaMaquinas)):
  if(listaMaquinas[i]["precio"] <= 300):
    contadorBaratas = contadorBaratas+1
  elif(listaMaquinas[i]["precio"] > 400):
    contadorCaras = contadorCaras+1
  i = i+1

print("Maquinas baratas: "+str(contadorBaratas))
print("Maquinas caras: "+str(contadorCaras))
```

**Acumuladores – Listas simples**

```python
edades = [5,20,35,16,12]

acumuladorEdades = 0
i = 0
while(i < len(edades)):
  acumuladorEdades = acumuladorEdades+edades[i]
  i = i+1

print("Suma edades: "+str(acumuladorEdades))
```

**Acumuladores – Listas simples - Solución**

```python
edades = [5,20,35,16,12]

acumuladorEdades = 0
contadorDe10a20 = 0
i = 0
while(i < len(edades)):
  if(edades[i]>= 10 and edades[i]<=20):
    contadorDe10a20 = contadorDe10a20+1
  acumuladorEdades = acumuladorEdades+edades[i]
  i = i+1

promedio = acumuladorEdades/len(edades)

print("Suma edades: "+str(acumuladorEdades))
print("Promedio de edades: "+str(promedio))
print("Personas entre 10 y 20: "+str(contadorDe10a20))
```
