**Acceder a valores de un diccionario**

```python
carro = {
  "marca":"Toyota",
  "modelo":"Toyota 4Runner SR5",
  "generación":2022,
  "color":"rojo"
}

print(carro["marca"])
print(type(carro["generación"]))
print(carro["color"])
```

**Modificar valores de un diccionario**

```python
carro = {
  "marca":"Toyota",
  "modelo":"Toyota 4Runner SR5",
  "generación":2022,
  "color":"negro"
}

print(carro["color"])
carro["color"] = "rojo"
print(carro["color"])
```

**Agregar valores a un diccionario**

```python
carro = {
  "marca":"Toyota",
  "modelo":"Toyota 4Runner SR5",
  "generación":2022,
  "color":"rojo"
}

carro["maletero"] = "SkyBox 16 Carbonite"
carro["guardabarros"] = "Pocket Style"
#print(carro["maletero"])
print(carro)
```

**Eliminar valores de un diccionario**

```python
carro = {
  "marca":"Toyota",
  "modelo":"Toyota 4Runner SR5",
  "generación":2022,
  "color":"rojo",
  "maletero":"SkyBox 16 Carbonite",
  "guardabarros":"Pocket Style"
}

del carro["guardabarros"]
print(carro)
```

**Listas con diccionarios**

```python
carro1 = {"marca":"Toyota","color":"rojo"}
carro2 = {"marca":"Mazda","color":"verde"}

listaCarros = []
listaCarros.append(carro1)
listaCarros.append(carro2)

print(listaCarros)
print(listaCarros[0])
print(listaCarros[0]["color"])
```

**Visualización en memoria de listas con diccionarios**

```python
carro1 = {"marca":"Toyota","color":"rojo"}
carro2 = {"marca":"Mazda","color":"verde"}

listaCarros = [carro1, carro2]
print(listaCarros[0]["color"])
```

**Ciclos con listas y diccionarios**

```python
carro1 = {"marca":"Toyota","color":"rojo"}
carro2 = {"marca":"Mazda","color":"verde"}

listaCarros = [carro1, carro2]

i = 0
while(i < len(listaCarros)):
  print("Marca es: "+listaCarros[i]["marca"])
  i = i+1
```

**Ejercicio de clase 1**
