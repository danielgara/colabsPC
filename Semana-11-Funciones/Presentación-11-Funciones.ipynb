**Funciones con retorno - Ejemplo**

```python
def calcularMulta(diasDeRetraso):
  if(diasDeRetraso > 7):
    return diasDeRetraso*800
  else:
    return 1000

dias = int(input("Ingrese los días de retraso: "))
deuda = calcularMulta(dias)
print("Debe pagar "+str(deuda)+" pesos de multa")
```

**Ejercicio de clase 2**

```python
def calcularMulta(diasDeRetraso, estrato):
  if(estrato == 1 or estrato == 2 or estrato == 3):
    return 0
  elif(diasDeRetraso > 5):
    return diasDeRetraso*900
  else:
    return 1200

dias = int(input("Ingrese los días de retraso: "))
estrato = int(input("Ingrese el estrato: "))
deuda = calcularMulta(dias, estrato)
print("Debe pagar "+str(deuda)+" pesos de multa")
```

**Mini proyecto librería**

```python
def mostrarLibros(libros):
  i = 0
  while(i < len(libros)):
    print("Nombre del libro "+str(i+1)+": ")
    print(libros[i])
    i = i+1

def agregarLibro(libros, nombre):
  libros.append(nombre)

libros = ["El Código da Vinci"]

while(True):
  print("----Ingrese----")
  print("0 para salir")
  print("1 para mostrar libros")
  print("2 para agregar un libro")
  operación = int(input())
  
  if(operación == 0):
    print("Chaito")
    break
  elif(operación == 1):
    mostrarLibros(libros)
  elif(operación == 2):
    nombre = input("Ingrese el nombre del libro: ")
    agregarLibro(libros, nombre)
```

**Mini proyecto librería – Retos**

```python
def mostrarLibros(libros):
  i = 0
  while(i < len(libros)):
    print("Nombre del libro "+str(i+1)+": ")
    print(libros[i])
    i = i+1

def agregarLibro(libros, nombre):
  libros.append(nombre)

def mostrarPrimerLibro(libros):
  print(libros[0])

def mostrarLibrosMayus(libros):
  i = 0
  while(i < len(libros)):
    print(libros[i].upper())
    i = i+1

def mostrarLibrosConAes(libros):
  i = 0
  while(i < len(libros)):
    if(libros[i].find("a") != -1):
      print(libros[i])
    i = i+1

def eliminarPos(libros, pos):
  libros.pop(pos)

def mostrarUltimoLibro(libros):
  if(len(libros) > 0):
    return libros[len(libros)-1]
  else:
    return "No hay libros"

libros = ["El Código da Vinci"]

while(True):
  print("----Ingrese----")
  print("0 para salir")
  print("1 para mostrar libros")
  print("2 para agregar un libro")
  print("3 para mostrar el primer libro")
  print("4 para mostrar libros en mayúscula")
  print("5 para mostrar libros con a")
  print("6 para eliminar un libro")
  print("7 para mostrar último libro")
  operación = int(input())
  
  if(operación == 0):
    print("Chaito")
    break
  elif(operación == 1):
    mostrarLibros(libros)
  elif(operación == 2):
    nombre = input("Ingrese el nombre del libro: ")
    agregarLibro(libros, nombre)
  elif(operación == 3):
    mostrarPrimerLibro(libros)
  elif(operación == 4):
    mostrarLibrosMayus(libros)
  elif(operación == 5):
    mostrarLibrosConAes(libros)
  elif(operación == 6):
    pos = int(input("Ingrese pos a eliminar: "))
    eliminarPos(libros, pos)
  elif(operación == 7):
    print(mostrarUltimoLibro(libros))
```
